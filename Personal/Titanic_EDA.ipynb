## The Data

I will be working with a famous titanic data set for the EDA.

```python
#importing usual packages
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Explaratory data analysis

```python
sns.jointplot(x='fare', y='age', data=titanic, color='darkblue')
plt.show()
```

```python
sns.histplot(x='fare', data=titanic, color = 'pink', bins=30)
plt.xlim(left=0)
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette={'First': 'lightblue', 'Second': 'lightgreen', 'Third': 'salmon'})
plt.show()
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
plt.show()
```

```python
sns.countplot(x='sex', data=titanic, palette = {'male': 'navy', 'female': 'green'})
plt.show()
```

```python
cr = titanic.corr(numeric_only=True)
sns.heatmap(cr, cmap='coolwarm')
plt.show()
```

```python
fg = sns.FacetGrid(col = 'sex', data = titanic)
fg.map(plt.hist, data=titanic, x='age')
plt.show()
```
